### Dependencies

```python
import numpy as np
import matplotlib.image as mpimg  # for reading in images

import matplotlib.pyplot as plt
import cv2  # computer vision library

%matplotlib inline
```

### Read and Display Image

```python
# Read in the image
image = mpimg.imread('../Data/WhiteTiger.jpg')

# Print out the image dimensions
print('Image dimensions:', image.shape)

# Change from color to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_image, cmap='gray');
```

```python
# Prints specific grayscale pixel values
# What is the pixel value at x = 250 and y = 575 (around the eye of the tiger)?

x = 250
y = 575

print(gray_image[y,x])
```

```python
# Finds the maximum and minimum grayscale values in this image

max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print(f'Max: {max_val}')
print(f'Min: {min_val}')
```

```python
# Close-up of the eye

plt.imshow(gray_image[600:675, 215:300], cmap="gray");
```

```python
# Close-up with numerical representation

img = np.squeeze(gray_image[600:675, 215:300])

fig = plt.figure(figsize = (20,20)) 
ax = fig.add_subplot(111)
ax.imshow(img, cmap='gray')
width, height = img.shape
thresh = img.max()/2.5
for x in range(width):
    for y in range(height):
        val = round(img[x][y],2) if img[x][y] !=0 else 0
        ax.annotate(str(val), xy=(y,x),
                    horizontalalignment='center',
                    verticalalignment='center',
                    color='white' if img[x][y]<thresh else 'black')
```
